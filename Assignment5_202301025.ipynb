Q.1 Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data from company_sales_data.csv
company_sales_data = pd.read_csv('company_sales_data.csv')

# Task 1: Read Total profit of all months and show it using a line plot
plt.figure(figsize=(10, 6))  # Set the figure size

sns.lineplot(data=company_sales_data['total_profit'], marker='o')  # Create line plot for total profit
plt.title('Total Profit of All Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

# Task 2: Read all product sales data and show it using a multiline plot
plt.figure(figsize=(10, 6))  # Set the figure size

# Create multiline plot for product sales
for product in company_sales_data.columns[1:7]:  # Assuming the columns 1-6 contain product sales data
    sns.lineplot(data=company_sales_data[product], label=product)

plt.title('Product Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Products', bbox_to_anchor=(1.05, 1), loc='upper left')  # Add legend outside the plot
plt.tight_layout()
plt.show()

# Task 3: Plot bar chart for all the features/attributes
plt.figure(figsize=(10, 6))  # Set the figure size

# Remove 'total_units' and 'total_profit' columns as they're not needed in the bar chart
attributes = company_sales_data.columns[1:7]

# Create bar chart for all features/attributes
sns.barplot(data=company_sales_data[attributes])
plt.title('Bar Chart for Features/Attributes')
plt.xlabel('Products')
plt.ylabel('Values')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

Q.2 Create a Scatter plot in Plotly (can be created using the scatter() method of
plotly.express.)

```python
import pandas as pd
import plotly.express as px

# Load data from company_sales_data.csv
company_sales_data = pd.read_csv('company_sales_data.csv')

# Create scatter plot
fig = px.scatter(company_sales_data, x='total_units', y='total_profit', title='Scatter Plot of Total Units vs Total Profit')

# Show the plot
fig.show()
```

Q.3 Create a Bar Chart in Plotly( can be created using the bar() method of
plotly.express class.)


```python
import pandas as pd
import plotly.express as px

# Load data from company_sales_data.csv
company_sales_data = pd.read_csv('company_sales_data.csv')

# Create bar chart
fig = px.bar(company_sales_data, x='month_number', y=['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer'],
             title='Product Sales by Month', labels={'value': 'Sales', 'month_number': 'Month'})

# Show the plot
fig.show()
```

Q.4 Create a histograms in plotly (histograms can be created using the
histogram() function of the plotly.express class.)

```python
import pandas as pd
import plotly.express as px

# Load data from company_sales_data.csv
company_sales_data = pd.read_csv('company_sales_data.csv')

# Create histograms for each product
fig = px.histogram(company_sales_data, x='total_units', title='Histogram of Total Units Sold',
                   labels={'total_units': 'Total Units'})

# Show the plot
fig.show()
```
